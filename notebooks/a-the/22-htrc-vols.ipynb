```python
import ujson
import bz2
import random
```

```python
class Volume:
    
    @classmethod
    def from_path(cls, path):
        with bz2.open(path) as fh:
            return cls(ujson.load(fh))
        
    def __init__(self, data):
        self.data = data
        
    def splat_tokens(self, p1, p2):
        
        tokens = []
        for page in self.data['features']['pages'][p1:p2+1]:
            
            page_tokens = []
            for token, pos_count in page['body']['tokenPosCount'].items():
                for pos, count in pos_count.items():
                    page_tokens += [token]*count
                    
            random.shuffle(page_tokens)
            tokens += page_tokens
            
        return tokens
```

```python
vol = Volume.from_path('../../data/htrc-vols/uc2.ark+=13960=t3qv3qv94.json.bz2')
```
