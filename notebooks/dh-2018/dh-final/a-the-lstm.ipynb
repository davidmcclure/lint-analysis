```python
def linspace_samples(df, s_low, s_high, n=50):

    rows = []
    for s in tqdm(np.linspace(s_low, s_high, n)):
        d = abs(df.score - s)
        rows.append(df.loc[d.sort_values().index].iloc[0])

    return pd.DataFrame(rows).sort_values('score', ascending=False)
```
