```python
import numpy as np
```

```python
np.random.multinomial(10000, [1/3, 1/3, 1/3])
```

```python
x, y = [], []
for n in range(10000, 10000000, 100000):
    
    print(n)
    
    ds = []
    for _ in range(1000):
        draw = np.random.multinomial(n, [1/3]*3)
        ds.append(draw[0]-draw[1])
        ds.append(draw[1]-draw[2])

    x.append(n)
    y.append(abs(np.array(ds)).mean())
    
x = np.array(x)
y = np.array(y)
```

```python
plt.plot(x, y)
plt.plot(x, np.sqrt(x*1/3*(2/3)))
```
