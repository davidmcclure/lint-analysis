```python
def plot_token_pos(token, min_count=1e5, w=8, h=8):
    
    pos_counts = BinCount.token_pos_counts(token)
    
    plt.figure(figsize=(w, h))
    
    lines = []
    for pos, count in pos_counts.items():
        if count > min_count:
            series = BinCount.token_series(token, pos=pos)
            line, = plt.plot(series / series.sum(), label=pos)
            lines.append(line)
        
    plt.legend(
        handles=lines,
        bbox_to_anchor=(1.03, 1),
        loc=2,
        borderaxespad=0,
    )
    
    plt.xlabel('Narrative time')
    plt.ylabel('Relative frequency')
    plt.show()
```

```python
plot_token_pos('that')
```

```python
def plot_tokens(tokens, w=8, h=8):
    
    plt.figure(figsize=(w, h))
    
    lines = []
    for token, pos in tokens:
        series = BinCount.token_series(token, pos=pos)
        line, = plt.plot(series / series.sum(), label=f'{token} ({pos})')
        lines.append(line)
        
    plt.legend(
        handles=lines,
        bbox_to_anchor=(1.03, 1),
        loc=2,
        borderaxespad=0,
    )
    
    plt.xlabel('Narrative time')
    plt.ylabel('Relative frequency')
    plt.show()
```
