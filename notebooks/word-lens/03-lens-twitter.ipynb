```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

mpl.style.use('seaborn-muted')
sns.set(style="whitegrid")

%matplotlib inline
```

```python
import pandas as pd
```

```python
novels = pd.read_json('/Users/dclure/Projects/data/word-len-by-novel.json/part-00000-9070510c-5cc5-4c2a-a275-f9fbecc0843d-c000.json', lines=True)
```

```python
novels.describe()
```

```python
novels_long = novels[novels['word_count'] > 50000]
```

```python
novels_long.describe()
```

```python
len(novels_long)
```

```python
plt.figure(figsize=(10, 6))
_ = plt.hist(novels_long['avg_word_len'].dropna(), bins=200)
```

```python
novels_long.sort_values('avg_word_len').head(20)
```

```python
novels_long.sort_values('avg_word_len', ascending=False).head(20)
```

```python
len(novels)
```

```python
grid = sns.jointplot(
    'pub_year',
    'avg_word_len',
    data=novels,
    kind='reg',
    size=15,
    ylim=(3, 4.5),
    ratio=8,
    marginal_kws=dict(bins=100),
    scatter_kws=dict(s=2),
)

grid.fig.set_figwidth(10)
grid.fig.set_figheight(10)
grid.fig.set_dpi(100)

plt.rc('axes', labelsize=16)
plt.rc('axes', titlesize=16)

plt.suptitle('Average word length, Gale American Fiction + Chicago (n=27166)', size=20, y=1.1)
plt.xlabel('Publication year')
plt.ylabel('Average characters per word')
```

```python
grid = sns.jointplot(
    'pub_year',
    'avg_word_len',
    data=novels_long[novels_long['corpus']=='gale'],
    kind='reg',
    size=15,
    ylim=(3, 4.5),
    ratio=8,
    marginal_kws=dict(bins=100),
    scatter_kws=dict(s=4),
)

grid.fig.set_figwidth(10)
grid.fig.set_figheight(10)

plt.rc('axes', labelsize=16)
plt.rc('axes', titlesize=16)

plt.xlabel('Publication year')
plt.ylabel('Average characters per word')
```
