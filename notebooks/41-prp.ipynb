```python
from lint_analysis.bin_counts.models import BinCount
```

```python
def plot_tokens(tokens, title=None, w=8, h=8):
    
    plt.figure(figsize=(w, h))
    
    lines = []
    for token, pos in tokens:
        series = BinCount.token_series(token, pos=pos)
        line, = plt.plot(series / series.sum(), label=f'{token} ({pos})')
        lines.append(line)
        
    plt.legend(
        handles=lines,
        bbox_to_anchor=(1.03, 1),
        loc=2,
        borderaxespad=0,
    )
    
    if title:
        plt.title(title)
    
    plt.xlabel('Narrative time')
    plt.ylabel('Relative frequency')
    plt.show()
```

# Subject

```python
plot_tokens([
    ('i', 'PRP'),
    ('you', 'PRP'),
    ('he', 'PRP'),
    ('she', 'PRP'),
    ('it', 'PRP'),
    ('we', 'PRP'),
    ('they', 'PRP'),
])
```

```python
plot_tokens([
    ('i', 'PRP'),
])
```

```python
plot_tokens([
    ('he', 'PRP'),
    ('she', 'PRP'),
])
```
