```python
import matplotlib as mpl
import matplotlib.pyplot as plt

%matplotlib inline
mpl.style.use('bmh')
```

```python
import numpy as np
```

```python
from lint_analysis.bin_counts import cache
```

```python
tokens = cache.token_counts(10000)
```

```python
data = []

for i, (token, count) in enumerate(tokens.items()):
    
    series = cache.token_series(token)
    exp_var = series.sum() * (1/100) * (1 - 1/100)
    
    data.append((count, series.var(), exp_var))
    
    if i % 1000 == 0: print(i)
```

```python
import pandas as pd

df = pd.DataFrame(
    data,
    columns=('count', 'var', 'exp'),
    index=tokens.keys(),
)
```

```python
pd.set_option('display.max_rows', 1000)

df.sort_values('var', ascending=False).head(100)
```

```python
plt.figure(figsize=(15, 10))

plt.xlabel('Word count')
plt.ylabel('Variance')
plt.xscale('log')
plt.yscale('log')
plt.scatter(df['count'], df['var'], s=2, c='b')
plt.plot(df['count'], df['exp'])

plt.show()
```

```python
import math

df_scored = df.copy()

df_scored['score'] = df.index.map(
    lambda t: df.loc[t]['var'] / df.loc[t]['exp']
)
```

```python
df_scored.sort_values('score', ascending=False).head(500)
```

```python
plt.figure(figsize=(15, 10))

plt.xlabel('Word count')
plt.ylabel('Variance')
plt.xscale('log')
plt.yscale('log')

norm = mpl.colors.LogNorm(
    vmin=df_scored['score'].min(),
    vmax=df_scored['score'].max(),
)

plt.scatter(
    df_scored['count'], df_scored['var'],
    s=7, c=df_scored['score'],
    cmap='spectral', norm=norm,
)

plt.plot(df['count'], df['exp'])

plt.show()
```

```python
def plot_series(token, y1=None, y2=None):
    
    series = cache.token_series(token, year1=y1, year2=y2)
    
    plt.figure(figsize=(10, 5))
    plt.bar(np.arange(len(series)), series)

    plt.title('"{}"'.format(token))
    plt.xlabel('Narrative time')
    plt.ylabel('Word count')
    plt.show()
```

```python
plot_series('gun')
plot_series('a')
```
