```python
import matplotlib as mpl
import matplotlib.pyplot as plt

%matplotlib inline
mpl.style.use('bmh')
```

```python
import numpy as np
```

```python
from lint_analysis.bin_counts.models import BinCount
```

```python
from joblib import Memory

mem = Memory('cache')

token_series = mem.cache(BinCount.token_series)
token_counts = mem.cache(BinCount.token_counts)
```

```python
tokens = token_counts(10000)
```

```python
data = []

for i, (token, count) in enumerate(tokens.items()):
    
    series = token_series(token)
    exp_var = series.sum() * (1/100) * (1 - 1/100)
    
    data.append((count, series.var(), exp_var))
    
    if i % 100 == 0: print(i)
```

```python
import pandas as pd

df = pd.DataFrame(data, columns=('count', 'var', 'exp'), index=tokens.keys())
```

```python
pd.set_option('display.max_rows', 1000)

df.sort_values('var', ascending=False).head(100)
```

```python
plt.figure(figsize=(15, 5))

plt.xlabel('Word count')
plt.ylabel('Variance')
plt.xscale('log')
plt.yscale('log')
plt.scatter(df['count'], df['var'], s=2, c='b')
plt.plot(df['count'], df['exp'])

plt.show()
```

```python
import math

df_scored = df.copy()

df_scored['score'] = df.index.map(
    lambda t: df.loc[t]['var'] / df.loc[t]['exp']
)
```

```python
df_scored['zscore'] = (
    (df_scored['score'] - df_scored['score'].mean()) /
    df_scored['score'].std(ddof=0)
)
```

```python
df_scored.sort_values('score', ascending=False).head(500)
```

```python
# https://matplotlib.org/devdocs/tutorials/colors/colormapnorms.html

class MidpointNormalize(mpl.colors.Normalize):
    
    def __init__(self, vmin=None, vmax=None, midpoint=None, clip=False):
        self.midpoint = midpoint
        mpl.colors.Normalize.__init__(self, vmin, vmax, clip)

    def __call__(self, value, clip=None):
        x, y = [self.vmin, self.midpoint, self.vmax], [0, 0.5, 1]
        return np.ma.masked_array(np.interp(value, x, y))
```
