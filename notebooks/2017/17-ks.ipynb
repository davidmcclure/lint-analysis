```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
from scipy import stats
```

```python
from lint_analysis.bin_counts.models import BinCount

def series_ks_2s(series):
    
    expected = series.sum() / len(series)
    uniform = np.full(len(series), expected)
    
    return stats.ks_2samp(series, uniform)
```

```python
series_ks_2s(BinCount.token_series('bla'))
```

```python
from joblib import Memory

mem = Memory('cache')

token_counts = mem.cache(BinCount.token_counts)

tokens = token_counts(10000)
```

```python
data = []
for i, (token, count) in enumerate(tokens.items()):
    
    series = BinCount.token_series(token)
    
    score = series_ks_2s(series)
    
    data.append((token, count, score))
    
    if i%100 == 0:
        print(i)
```

```python
import pandas as pd

df = pd.DataFrame(data, columns=('token', 'count', 'score'))
```

```python
pd.set_option('display.max_rows', 1000)

df.sort_values('score', ascending=False).head(100)
```

```python
x = df['count']
y = df['score'].map(lambda r: r[0])
```

```python
plt.figure(figsize=(15, 5))

plt.xlabel('Word count')
plt.ylabel('Kolmogorov-Smirnov')
plt.xscale('log')
plt.scatter(x, y, s=2, c='b')

plt.show()
```
