```python
import matplotlib as mpl
import matplotlib.pyplot as plt

%matplotlib inline
mpl.style.use('bmh')
```

```python
from lint_analysis.bin_counts.cache import build_variance_df
```

```python
df = build_variance_df()
```

```python
uni_x = [df['count'].min(), df['count'].max()]
uni_y = [n * (1/100) * (1 - 1/100) for n in uni_x]
```

```python
plt.figure(figsize=(15, 5))
plt.xscale('log')
plt.yscale('log')
plt.scatter(df['count'], df['variance'], s=2, c='b')
plt.plot(uni_x, uni_y)
plt.show()
```

```python
data = []

for token, row in df.iterrows():
    expected = row['count'] * (1/100) * (1 - 1/100)
    score = row['variance'] / expected
    data.append((row['count'], row['variance'], score))
```

```python
import pandas as pd

scores = pd.DataFrame(
    data,
    columns=('count', 'variance', 'score'),
    index=df.index,
)
```

```python
pd.set_option('display.max_rows', 1000)

scores.sort_values('score', ascending=False)
```
