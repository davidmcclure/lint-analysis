```python
from numpy.random import multinomial

def sample_series(series, n=1000):
    return multinomial(series.sum(), series/series.sum(), n)
```

```python
def zscore_series(series):
    
    expected = series.sum() / len(series)
    samples = sample_series(np.full(len(series), expected))
    
    avg_std = samples.std(1).mean()
    
    return (series - expected) / avg_std
```

```python
import numpy as np

from lint_analysis.bin_counts.models import BinCount

def plot_zscore_series(token, w=5, h=5):
    
    series = BinCount.token_series(token)
    
    zscores = zscore_series(series)

    plt.figure(figsize=(w,h))
    
    print(token)
    
    plt.title(token)
    plt.xlabel('Narrative time')
    plt.xlabel('Word count')
    plt.plot(zscores)
    plt.show()
```

```python
from lint_analysis.bin_counts.models import BinCount

def plot_cohort(*tokens, b1=0, b2=100, w=8, h=8):
    
    plt.figure(figsize=(w, h))
    
    lines = []
    for token in tokens:
        
        series = zscore_series(BinCount.token_series(token))
        line, = plt.plot(series[b1:b2], label=token)
        lines.append(line)
        
    plt.legend(
        handles=lines,
        bbox_to_anchor=(1.03, 1),
        loc=2,
        borderaxespad=0,
    )
    
    plt.xlabel('Narrative time')
    plt.ylabel('Relative frequency')
    plt.show()
```

```python
def plot_cohort_colors(*tokens, b1=0, b2=100, w=8, h=8):
    
    plt.figure(figsize=(w, h))
    
    lines = []
    for token, color in tokens:
        
        series = zscore_series(BinCount.token_series(token))
        line, = plt.plot(series[b1:b2], label=token, c=color)
        lines.append(line)
        
    plt.legend(
        handles=lines,
        bbox_to_anchor=(1.03, 1),
        loc=2,
        borderaxespad=0,
    )
    
    plt.xlabel('Narrative time')
    plt.ylabel('Relative frequency')
    plt.show()
```
