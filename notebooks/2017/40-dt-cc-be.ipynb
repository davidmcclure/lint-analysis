```python
def plot_tokens(tokens, title=None, w=8, h=8):
    
    plt.figure(figsize=(w, h))
    
    lines = []
    for token, pos in tokens:
        series = cache.token_series(token, pos=pos)
        line, = plt.plot(series / series.sum(), label=f'{token} ({pos})')
        lines.append(line)
        
    plt.legend(
        handles=lines,
        bbox_to_anchor=(1.03, 1),
        loc=2,
        borderaxespad=0,
    )
    
    if title:
        plt.title(title)
    
    plt.xlabel('Narrative time')
    plt.ylabel('Relative frequency')
    plt.show()
```

```python
import math
import numpy as np

def plot_token(token, pos, w=8, h=8):
    
    series = cache.token_series(token, pos=pos)
    
    n = series.sum()
    p = 1 / len(series)
    
    mean = n * p
    var = mean * (1-p)
    std = math.sqrt(var)

    plt.figure(figsize=(w,h))
    
    plt.plot([0, 100], [mean, mean], c='orange', lw=0.5)

    plt.fill_between(
        range(101),
        mean + 2*std,
        mean - 2*std,
        color='black',
        alpha=0.1,
    )
    
    plt.title(f'{token} ({pos})')
    plt.xlabel('Narrative time')
    plt.xlabel('Word count')
    plt.plot(series)
    plt.show()
```

```python
import math
import numpy as np

def plot_poss(poss, w=8, h=8):
    
    plt.figure(figsize=(w,h))
    
    lines = []
    for pos in poss:
        series = cache.pos_series(pos)
        line, = plt.plot(series / series.sum(), label=pos)
        lines.append(line)
        
    plt.legend(
        handles=lines,
        bbox_to_anchor=(1.03, 1),
        loc=2,
        borderaxespad=0,
    )
    
    plt.title(', '.join(poss))
    plt.xlabel('Narrative time')
    plt.xlabel('Word count')
    plt.show()
```
