```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np

from numpy.random import multinomial

def sample_100_bins(count, nsamples=10000):
    return multinomial(count, np.full(100, 0.01), nsamples)
```

```python
x = np.array(range(0, 1000000, 10000))
y = []

for count in x:
    s = sample_100_bins(count)
    avg_std = s.var(0).mean()
    y.append(avg_std)
    
plt.figure(figsize=(20, 10))
plt.plot(x, y)
plt.plot(x, x/100)
plt.show()
```
