```python
import math
import numpy as np

def plot_raw_pos(pos, label, w=4, h=4):
    
    plt.figure(figsize=(w,h))
    
    series = cache.pos_series(pos)
    
    mean = series.sum() / len(series)
    plt.axhline(mean, c='black', lw=0.5, alpha=0.5)
    
#     plt.plot(series)

    plt.hist(
        np.arange(len(series)),
        weights=series,
        bins=len(series),
        zorder=3,
    )
        
    plt.title(f'{pos} ({label})')
    plt.xlabel('Narrative time')
    plt.xlabel('Word count')
    plt.show()
```

```python
for tag, label in tags.items():
    series = cache.pos_series(tag)
    print(series.sum())
    plot_raw_pos(tag, label)
```
