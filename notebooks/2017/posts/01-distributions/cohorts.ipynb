# Cohorts

```python
import matplotlib as mpl
import matplotlib.pyplot as plt

%matplotlib inline
mpl.style.use('bmh')
```

```python
from lint_analysis.bin_counts import cache
```

```python
def plot_cohort(title, tokens, w=8, h=8):
    
    plt.figure(figsize=(w, h))
    
    lines = []
    for token in tokens:
        
        series = cache.token_series(token)
        line, = plt.plot(series, label=token)
        lines.append(line)
        
    plt.legend(
        handles=lines,
        bbox_to_anchor=(1.03, 1),
        loc=2,
        borderaxespad=0,
    )
    
    plt.title(title)
    plt.xlabel('Narrative time')
    plt.ylabel('Relative frequency')
    plt.show()
```

```python
def plot_cohort_colors(title, tokens, w=8, h=8):
    
    plt.figure(figsize=(w, h))
    
    lines = []
    for token, color in tokens:
        
        series = cache.token_series(token)
        line, = plt.plot(series, label=token, c=color)
        lines.append(line)
        
    plt.legend(
        handles=lines,
        bbox_to_anchor=(1.03, 1),
        loc=2,
        borderaxespad=0,
    )
    
    plt.title(title)
    plt.xlabel('Narrative time')
    plt.ylabel('Relative frequency')
    plt.show()
```
