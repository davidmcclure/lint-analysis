```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

%matplotlib inline
mpl.style.use('bmh')
```

```python
from lint_analysis.bin_counts import cache
```

```python
from lint_analysis.bin_counts.models import BinCount
```

```python
def plot_tokens(tokens, title=None, w=5, h=5):
    
    plt.figure(figsize=(w, h))
    
    lines = []
    for token, pos in tokens:
        series = cache.token_series(token, pos=pos)
        line, = plt.plot(series / series.sum(), label=f'{token} ({pos})')
        lines.append(line)
        
    plt.legend(
        handles=lines,
        bbox_to_anchor=(1.03, 1),
        loc=2,
        borderaxespad=0,
    )
    
    if title:
        plt.title(title)
    
    plt.xlabel('Narrative time')
    plt.ylabel('Relative frequency')
    plt.show()
```

```python
plot_tokens([
    ('he', 'PRP'),
    ('she', 'PRP'),
])
```
