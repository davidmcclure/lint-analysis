# Variance

```python
import matplotlib as mpl
import matplotlib.pyplot as plt

%matplotlib inline
mpl.style.use('bmh')
```

```python
import numpy as np
```

```python
from lint_analysis.bin_counts import cache
```

```python
tokens = cache.token_counts(10000)
```

```python
data = []

for i, (token, count) in enumerate(tokens.items()):
    
    series = cache.token_series(token)
    exp_var = series.sum() * (1/100) * (1 - 1/100)
    
    data.append((count, series.var(), exp_var))
    
    if i % 1000 == 0: print(i)
```

```python
import pandas as pd

df = pd.DataFrame(
    data,
    columns=('count', 'var', 'exp'),
    index=tokens.keys(),
)
```

```python
pd.set_option('display.max_rows', 1000)

df.sort_values('var', ascending=False).head(100)
```

```python
plt.figure(figsize=(10, 8))

plt.title('Variance, linear scales')
plt.xlabel('Word count')
plt.ylabel('Variance')
plt.scatter(df['count'], df['var'], s=3, c='b')

plt.show()
```

```python
plt.figure(figsize=(10, 8))

plt.title('Variance, log-log')
plt.xlabel('Word count')
plt.ylabel('Variance')
plt.xscale('log')
plt.yscale('log')
plt.scatter(df['count'], df['var'], s=3, c='b')

plt.show()
```

```python
plt.figure(figsize=(10, 8))

plt.title('Observed vs. expected variance')
plt.xlabel('Word count')
plt.ylabel('Variance')
plt.xscale('log')
plt.yscale('log')
plt.scatter(df['count'], df['var'], s=3, c='b')
plt.plot(df['count'], df['exp'])

plt.show()
```

```python
import math

df['score'] = df.index.map(
    lambda t: df.loc[t]['var'] / df.loc[t]['exp']
)
```

```python
df.sort_values('score', ascending=False).head(100)
```

```python
plt.figure(figsize=(10, 7))

plt.title('Observed / expected ratios')
plt.xlabel('Word count')
plt.ylabel('Variance')
plt.xscale('log')
plt.yscale('log')

norm = mpl.colors.LogNorm(
    vmin=df['score'].min(),
    vmax=df['score'].max(),
)

plt.scatter(
    df['count'], df['var'],
    s=3, c=df['score'],
    cmap='spectral', norm=norm,
)

plt.plot(df['count'], df['exp'])

plt.show()
```
