```python
def plot_tokens(tokens, title=None, w=5, h=5):
    
    plt.figure(figsize=(w, h))
    
    lines = []
    for token, pos in tokens:
        series = cache.token_series(token, pos=pos)
        line, = plt.plot(series / series.sum(), label=f'{token} ({pos})')
        lines.append(line)
        
    plt.legend(
        handles=lines,
        bbox_to_anchor=(1.03, 1),
        loc=2,
        borderaxespad=0,
    )
    
    if title:
        plt.title(title)
    
    plt.xlabel('Narrative time')
    plt.ylabel('Relative frequency')
    plt.show()
```

```python
import math
import numpy as np

def plot_poss(poss, w=5, h=5):
    
    plt.figure(figsize=(w,h))
    
    lines = []
    for pos in poss:
        series = cache.pos_series(pos)
        line, = plt.plot(series / series.sum(), label=pos)
        lines.append(line)
        
    plt.legend(
        handles=lines,
        bbox_to_anchor=(1.03, 1),
        loc=2,
        borderaxespad=0,
    )
    
    plt.title(', '.join(poss))
    plt.xlabel('Narrative time')
    plt.xlabel('Relative frequency')
    plt.show()
```

```python
plt.figure(figsize=(5,5))

poss = sum([
    cache.pos_series(pos)
    for pos in ['NN', 'NNS']
])

lines = []

line, = plt.plot(poss / poss.sum(), label='Nouns (NN + NNS)')
lines.append(line)

the = cache.token_series('the')
line, = plt.plot(the / the.sum(), label='the')
lines.append(line)
        
plt.legend(
    handles=lines,
    bbox_to_anchor=(1.03, 1),
    loc=2,
    borderaxespad=0,
)

plt.title('The vs. nouns')
plt.xlabel('Narrative time')
plt.xlabel('Word count')
plt.show()
```
