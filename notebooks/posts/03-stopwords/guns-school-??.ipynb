```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

%matplotlib inline
mpl.style.use('bmh')
```

```python
from lint_analysis.bin_counts import cache
```

```python
def plot_tokens(tokens, title=None, w=4, h=4):
    
    plt.figure(figsize=(w, h))
    
    lines = []
    for token in tokens:
        series = cache.token_series(token)
        line, = plt.plot(series / series.sum(), label=token)
        lines.append(line)
        
    plt.legend(
        handles=lines,
        bbox_to_anchor=(1.03, 1),
        loc=2,
        borderaxespad=0,
    )
    
    if title:
        plt.title(title)
    
    plt.xlabel('Narrative time')
    plt.ylabel('Relative frequency')
    plt.show()
```

```python
plot_tokens(['bullet', 'pistol', 'gun'], 'Guns')
plot_tokens(['student', 'students', 'school'], 'School')
plot_tokens(['called', 'these', 'a', 'an', 'people', 'or', 'house', 'two', 'man', 'than'], '???')
```
